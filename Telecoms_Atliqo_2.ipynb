```python
df_city = pd.read_csv("dim_cities.csv")

df_date = pd.read_csv("dim_date.csv")

df_plan= pd.read_csv("dim_plan.csv")

df_atliqo = pd.read_csv("fact_atliqo_metrics.csv")

df_mark_share = pd.read_csv("fact_market_share.csv")

df_revenue = pd.read_csv("fact_plan_revenue.csv")


df_metrics = pd.read_excel("metrics_list.xlsx")
```

## PREPARE PLANS TABLE

```python
df_revenue.head()
```

There are 120 unique "date" and "city_code" pairs which is equal to the rows of df_atliqo.

**Merge with City**

```python
df_rev_city = df_revenue.merge(df_city, how = 'left', on = 'city_code')
```

**Merge with date**

```python

df_rev_city_date = df_rev_city.merge(df_date, how = 'left', on = 'date')
```

**Merge with plan descriptions**

```python
df_rev_city_date_plan = df_rev_city_date.merge(df_plan, how = 'left', left_on = 'plans', right_on = 'plan')
```

```python
# Dropping the plan column as it is redundant
df_rev_city_date_plan.drop(columns = 'plan', inplace = True)
```

## Prepare Total Data

```python
df_atliqo.head()
```

**Merge with City**

```python
df_atliqo_city = df_atliqo.merge(df_city, how = 'left', on = 'city_code')  
```

**Merge with Date**

```python

df_atliqo_city_date = df_atliqo_city.merge(df_date, how = 'left', on = 'date')
```

```python
df_atliqo_city_date.head()
```

### Feature Engineering

```python
df_atliqo_city_date["revenue_lost_crores"] = df_atliqo_city_date["unsubscribed_users_lakhs"]* df_atliqo_city_date["arpu"]/100
```

```python
df_atliqo_city_date["perc_revenue_lost"] = (df_atliqo_city_date["revenue_lost_crores"] / df_atliqo_city_date["atliqo_revenue_crores"])*100
```

```python
df_atliqo_city_date.tail()
```

```python
df_atliqo_city_date.to_csv('atliqo_total.csv', index=False) 
```

## Market Share Dataset

```python
df_mark_share.head()
```

```python
df_mark_share_city =  df_mark_share.merge(df_city, how = 'left', on = ['city_code']) 
```

```python
df_mark_share_city
```

```python
df_mark_share_city_period = df_mark_share_city.merge(df_date, how = 'left', on = ['date']) 
```

```python
df_mark_share_city_period
```

```python
#df_mark_share_city_period.to_csv('atliqo_mrkt_share.csv', index=False) 
```

```python
df_mark_share_city_period["Revenue_crores"] = (df_mark_share_city_period["ms_pct"]/100)*(df_mark_share_city_period["tmv_city_crores"])
```

```python
df_mark_share_city_period.head()
```

```python
df_total_market = (df_mark_share_city_period.groupby(["before/after_5g","company"])["Revenue_crores"].sum()/df_mark_share_city_period.groupby("before/after_5g")["Revenue_crores"].sum()*100).reset_index()
df_total_market 
```

```python
#df_total_market.to_csv('total_market_share.csv', index = False)
```

```python
df_plan_contribution= (df_rev_city_date_plan.groupby(['before/after_5g','plans'])["plan_revenue_crores"].sum()/df_rev_city_date_plan.groupby('before/after_5g')["plan_revenue_crores"].sum()*100).reset_index()

df_plan_contribution
```
